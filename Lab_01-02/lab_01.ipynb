## 1. Problem „czarnej skrzynki”

W ostatnich latach systemy oparte na uczeniu maszynowym zaczęły być wykorzystywane w coraz większej liczbie obszarów naszego życia. Modele predykcyjne są stosowane między innymi w medycynie, finansach, marketingu, analizie ryzyka, systemach rekomendacyjnych, a także w autonomicznych pojazdach. W wielu przypadkach ich skuteczność przewyższa tradycyjne metody statystyczne, a czasem nawet decyzje podejmowane przez ludzi.

Jednak wraz z rosnącą złożonością modeli pojawia się poważny problem. Wiele współczesnych algorytmów działa w sposób, który dla człowieka jest trudny do zrozumienia. Modele takie jak lasy losowe, gradient boosting czy sieci neuronowe potrafią analizować ogromne ilości danych i wykrywać złożone zależności, ale jednocześnie ich wewnętrzne działanie jest często nieprzejrzyste. Oznacza to, że jesteśmy w stanie uzyskać bardzo dobrą predykcję, ale nie zawsze wiemy, dlaczego model podjął konkretną decyzję.

Taki model często określa się mianem **czarnej skrzynki** (ang. *black box*). Oznacza to system, którego działanie obserwujemy tylko poprzez wejścia i wyjścia. Wprowadzamy dane, otrzymujemy wynik, ale mechanizm prowadzący do tej decyzji pozostaje niejasny.

Wyobraźmy sobie prosty przykład. Bank korzysta z modelu uczenia maszynowego do oceny wniosków kredytowych. Klient składa wniosek o kredyt, a system automatycznie podejmuje decyzję: kredyt przyznany lub odrzucony. Model może mieć bardzo wysoką skuteczność – na przykład 95% poprawnych decyzji – jednak dla klienta i pracownika banku kluczowe pytanie brzmi: **dlaczego?**

Dlaczego jeden wniosek został zaakceptowany, a inny odrzucony? Czy decydujący był dochód, historia kredytowa, wiek, a może jakiś zupełnie inny czynnik? Jeśli nie jesteśmy w stanie tego wyjaśnić, pojawia się problem zaufania do systemu.

Właśnie w tym miejscu pojawia się koncepcja **Explainable Artificial Intelligence**, czyli wyjaśnialnej sztucznej inteligencji.

# 2. Czym jest XAI

Termin **Explainable Artificial Intelligence**, w skrócie **XAI**, odnosi się do zbioru metod, narzędzi i podejść, których celem jest umożliwienie człowiekowi zrozumienia działania modeli uczenia maszynowego.

Mówiąc prościej, XAI odpowiada na pytanie:

**dlaczego model podjął taką, a nie inną decyzję?**

Wyjaśnienia mogą przyjmować różne formy. Na przykład:

* wskazanie, które cechy miały największy wpływ na decyzję modelu,
* pokazanie, jak zmieniłaby się predykcja, gdyby zmieniła się wartość danej cechy,
* wygenerowanie reguł opisujących działanie modelu,
* wizualizacja zależności między cechami a predykcją.

Celem XAI nie jest koniecznie pełne zrozumienie wszystkich obliczeń wykonywanych przez model. W wielu przypadkach byłoby to niemożliwe lub niepraktyczne. Zamiast tego dąży się do uzyskania wyjaśnień, które są **wystarczająco zrozumiałe dla człowieka**, aby móc ocenić, czy model działa w sposób sensowny i wiarygodny.

## 2.2. Kluczowe pojęcia w praktyce

W literaturze dotyczącej XAI pojawia się wiele pojęć, które mogą być interpretowane na różne sposoby. W praktyce warto skupić się na kilku najważniejszych koncepcjach.

---

### Wyjaśnienie globalne

Wyjaśnienia globalne opisują **ogólne zachowanie modelu**.

Odpowiadają na pytania takie jak:

* które cechy są ogólnie najważniejsze dla modelu,
* jak zmienia się predykcja modelu w zależności od wartości danej cechy,
* jakie ogólne wzorce można zaobserwować w działaniu modelu.

Przykładem globalnego wyjaśnienia jest ranking ważności cech lub wykres zależności częściowych (PDP).

---

### Wyjaśnienie lokalne

Wyjaśnienia lokalne koncentrują się na **jednej konkretnej obserwacji**.

Odpowiadają na pytanie:

dlaczego model podjął taką decyzję w tym konkretnym przypadku?

Na przykład dla jednego wniosku kredytowego model może wskazać, że największy wpływ na decyzję miały:

* wysoki poziom zadłużenia,
* niska stabilność zatrudnienia,
* krótka historia kredytowa.

Do generowania wyjaśnień lokalnych często stosuje się metody takie jak LIME lub SHAP.

---

## 2.3. Dwa typy wyjaśnień

Nie wszystkie wyjaśnienia modeli są jednakowe. W literaturze często rozróżnia się dwa ważne typy wyjaśnień.

---

### Wyjaśnienia wierne modelowi (faithful)

Wyjaśnienie wierne modelowi oznacza, że dobrze odzwierciedla ono rzeczywiste działanie modelu.

Innymi słowy, jeśli wyjaśnienie mówi, że dana cecha jest bardzo ważna, to rzeczywiście zmiana tej cechy powinna istotnie wpływać na predykcję modelu.

Przykładem metody, która jest stosunkowo wierna modelowi, jest **permutation feature importance**. Polega ona na losowym mieszaniu wartości danej cechy i obserwowaniu, jak bardzo pogarsza się jakość modelu.

Jeżeli po permutacji danej cechy model działa znacznie gorzej, oznacza to, że cecha była ważna dla predykcji.

---

### Wyjaśnienia przekonujące dla człowieka (plausible)

Niektóre wyjaśnienia są zaprojektowane przede wszystkim po to, aby były **zrozumiałe dla człowieka**.

Nie zawsze oznacza to jednak, że dokładnie odzwierciedlają działanie modelu.

Na przykład metoda LIME buduje lokalny model liniowy w pobliżu analizowanej obserwacji. Model ten jest łatwy do interpretacji, ale jest tylko przybliżeniem rzeczywistego modelu.

W niektórych przypadkach może to prowadzić do tzw. **konfabulacji**, czyli sytuacji, w której wyjaśnienie brzmi przekonująco, ale nie oddaje w pełni logiki modelu.

---

## 2.4. Najczęstsze pułapki w interpretacji wyjaśnień

Analiza wyjaśnień modeli wymaga ostrożności, ponieważ łatwo jest popełnić błędy interpretacyjne.

---

### Korelacja cech i proxy

Jednym z najczęstszych problemów jest występowanie cech silnie ze sobą skorelowanych.

Jeśli dwie cechy są bardzo podobne, model może używać jednej z nich jako zastępnika drugiej. W takiej sytuacji interpretacja ważności cech może być myląca.

Na przykład model oceniający ryzyko kredytowe może wykorzystywać kod pocztowy jako przybliżenie poziomu dochodów w danym regionie.

---

### Wyjaśnienie nie oznacza przyczynowości

Bardzo ważne jest zrozumienie, że większość metod XAI opisuje **zależności statystyczne**, a nie związki przyczynowe.

Jeśli model pokazuje, że dana cecha jest silnie powiązana z wynikiem predykcji, nie oznacza to automatycznie, że ta cecha powoduje dany wynik.

Może istnieć ukryta zmienna, która wpływa zarówno na cechę, jak i na wynik.

---

### Niestabilność wyjaśnień lokalnych

Niektóre metody lokalne są wrażliwe na drobne zmiany danych.

Jeśli zmienimy nieznacznie wartości cech lub losowość procesu generowania próbek, wyjaśnienie może się zmienić.

Dlatego w praktyce często analizuje się stabilność wyjaśnień, aby upewnić się, że nie są one przypadkowe.

---

### Wysoka jakość modelu nie oznacza dobrej kalibracji

Model może osiągać bardzo wysokie wyniki na metrykach takich jak AUC czy accuracy, ale jednocześnie generować **źle skalibrowane prawdopodobieństwa**.

Oznacza to, że model potrafi poprawnie klasyfikować obserwacje, ale jego przewidywane prawdopodobieństwa nie odpowiadają rzeczywistej częstości zdarzeń.

W kontekście XAI jest to ważne, ponieważ interpretacja decyzji modelu często opiera się na analizie prawdopodobieństw predykcji.

---

# 3. Interpretowalność a wyjaśnialność

W kontekście XAI bardzo często pojawiają się dwa pojęcia, które bywają mylone: **interpretowalność** oraz **wyjaśnialność**.

Interpretowalność odnosi się do modeli, które są zrozumiałe same w sobie. Oznacza to, że człowiek może bezpośrednio przeanalizować strukturę modelu i zrozumieć jego działanie.

Przykładem takiego modelu jest regresja liniowa. Jeśli model ma postać:

cena = 2000 + 100 × metraż

to interpretacja jest prosta. Każdy dodatkowy metr kwadratowy zwiększa cenę mieszkania o 100 jednostek. Współczynniki modelu mają bezpośrednie znaczenie i można je łatwo wyjaśnić.

Innym przykładem interpretowalnego modelu jest drzewo decyzyjne. Drzewo składa się z zestawu reguł typu „jeśli – to”. Na przykład:

jeśli dochód > 5000 i historia kredytowa jest dobra, to przyznaj kredyt.

Taka struktura jest stosunkowo łatwa do zrozumienia nawet dla osoby, która nie zajmuje się uczeniem maszynowym.

Z kolei wyjaśnialność odnosi się do sytuacji, w której model sam w sobie nie jest łatwy do interpretacji, ale można zastosować dodatkowe metody, które pomagają zrozumieć jego decyzje. Dotyczy to przede wszystkim bardziej złożonych modeli, takich jak lasy losowe, modele boostingowe czy sieci neuronowe.

W takich przypadkach wykorzystuje się metody wyjaśniające działanie modelu po jego wytrenowaniu. Metody te nazywane są często **metodami post-hoc**, ponieważ analiza następuje po zbudowaniu modelu.

## Interpretowalność

Interpretowalność oznacza, że model jest **z natury zrozumiały dla człowieka**. Oznacza to, że możemy bezpośrednio przeanalizować strukturę modelu i zrozumieć jego działanie.

Przykładem jest regresja logistyczna. W tym modelu każda cecha ma przypisany współczynnik, który określa jej wpływ na wynik predykcji.

Jeżeli współczynnik dla danej cechy jest dodatni, oznacza to, że wzrost tej cechy zwiększa prawdopodobieństwo określonego wyniku. Jeżeli współczynnik jest ujemny, wpływ jest odwrotny.

Podobnie interpretowalne są niewielkie drzewa decyzyjne, w których każda decyzja jest podejmowana na podstawie prostej reguły.

---

## Wyjaśnialność

Wyjaśnialność odnosi się do sytuacji, w której model sam w sobie nie jest łatwy do zrozumienia, ale można zastosować dodatkowe metody, które pomagają wyjaśnić jego działanie.

Przykładem może być las losowy lub model gradient boosting. Tego typu modele składają się z wielu komponentów i trudno jest bezpośrednio prześledzić ich proces decyzyjny.

Dlatego stosuje się metody takie jak:

* permutation feature importance,
* LIME,
* SHAP.

Metody te generują wyjaśnienia na podstawie analizy zachowania modelu.

---

# 4. Dlaczego wyjaśnialność jest ważna

Można zadać pytanie: jeśli model działa dobrze i osiąga wysoką skuteczność, czy naprawdę musimy wiedzieć, dlaczego działa? W wielu przypadkach odpowiedź brzmi: tak.

Pierwszym powodem jest **bezpieczeństwo**. W systemach medycznych, finansowych czy transportowych decyzje podejmowane przez algorytmy mogą mieć poważne konsekwencje dla ludzi. Lekarz korzystający z systemu diagnostycznego musi wiedzieć, na jakiej podstawie model sugeruje określoną diagnozę. Jeśli system wskazuje, że pacjent ma wysokie ryzyko choroby, lekarz powinien móc zobaczyć, które czynniki miały największy wpływ na tę ocenę.

Drugim powodem są **regulacje prawne**. W Unii Europejskiej obowiązują przepisy dotyczące ochrony danych osobowych, w tym rozporządzenie GDPR. W pewnych sytuacjach osoby, których dotyczą automatyczne decyzje podejmowane przez algorytmy, mają prawo do uzyskania informacji o logice stojącej za tymi decyzjami. Oznacza to, że organizacje wykorzystujące modele uczenia maszynowego muszą być w stanie w pewnym stopniu wyjaśnić ich działanie.

Trzecim powodem jest **wykrywanie uprzedzeń i błędów w danych**. Modele uczą się na podstawie danych historycznych. Jeśli dane te zawierają błędy lub odzwierciedlają istniejące uprzedzenia społeczne, model może je nieświadomie powielać. Znane są przypadki systemów rekrutacyjnych, które faworyzowały określone grupy kandydatów, ponieważ dane treningowe były nierównomierne.

Czwartym powodem jest **debugowanie modeli**. Wyjaśnienia mogą pomóc odkryć problemy takie jak przeciek informacji między danymi treningowymi i testowymi, nadmierne dopasowanie modelu do danych czy nieprawidłowe przygotowanie cech.

## 4.1. Kto potrzebuje wyjaśnień?

Jednym z najważniejszych pytań w kontekście wyjaśnialnej sztucznej inteligencji jest to, **dla kogo właściwie tworzymy wyjaśnienia**. Różne grupy użytkowników mają różne potrzeby informacyjne i oczekują innych typów wyjaśnień. W praktyce oznacza to, że jedno wyjaśnienie modelu może być użyteczne dla inżyniera, ale zupełnie niezrozumiałe dla osoby, której dotyczy decyzja modelu.

Dlatego w projektowaniu systemów XAI kluczowe jest zidentyfikowanie **interesariuszy systemu**.

---

### Użytkownik lub beneficjent decyzji modelu

Pierwszą i często najważniejszą grupą jest **osoba, której dotyczy predykcja modelu**. Może to być klient banku, pacjent, kandydat do pracy lub użytkownik platformy internetowej.

Dla takiej osoby najważniejsze są pytania bardzo praktyczne:

* dlaczego model podjął taką decyzję wobec mnie?
* które czynniki miały największy wpływ na tę decyzję?
* czy mogę coś zmienić, aby w przyszłości decyzja była inna?

Wyobraźmy sobie system oceny wniosków kredytowych. Jeśli klient otrzymuje informację, że kredyt został odrzucony, naturalną reakcją jest pytanie: **dlaczego?**

Wyjaśnienie może wyglądać na przykład tak:

* niski dochód w stosunku do kwoty kredytu,
* krótka historia kredytowa,
* wysoki poziom istniejącego zadłużenia.

Takie informacje są dla użytkownika zrozumiałe i pozwalają mu podjąć działania w przyszłości, na przykład poprawić historię kredytową lub zmniejszyć poziom zadłużenia.

---

### Ekspert domenowy

Drugą ważną grupą są **eksperci z danej dziedziny**, czyli osoby posiadające wiedzę specjalistyczną niezwiązaną bezpośrednio z uczeniem maszynowym.

Przykłady takich ekspertów to:

* lekarze i diagnostycy w systemach medycznych,
* psychologowie kliniczni w systemach wspomagających diagnozę,
* analitycy finansowi w systemach oceny ryzyka,
* specjaliści HR w systemach rekrutacyjnych.

Ekspert domenowy niekoniecznie musi rozumieć szczegóły algorytmu, ale musi być w stanie odpowiedzieć na pytanie:

**czy model podejmuje decyzje zgodne z wiedzą ekspercką?**

Przykładowo w medycynie model może przewidywać ryzyko choroby na podstawie wielu cech pacjenta. Lekarz będzie chciał sprawdzić, czy model rzeczywiście opiera się na sensownych czynnikach, takich jak wiek pacjenta, wyniki badań czy styl życia, a nie na przypadkowych korelacjach.

Jeżeli model okaże się opierać swoje decyzje na cechach, które z punktu widzenia medycyny są nieistotne, może to oznaczać problem z danymi lub procesem treningu modelu.

---

### Inżynier lub data scientist

Kolejną grupą są osoby odpowiedzialne za **projektowanie, trenowanie i utrzymanie modeli**.

Dla inżyniera wyjaśnienia modeli są niezwykle ważnym narzędziem diagnostycznym. Pozwalają one między innymi:

* zidentyfikować błędy w danych treningowych,
* wykryć przeciek informacji (data leakage),
* sprawdzić, czy model nie uczy się niepożądanych zależności,
* monitorować zmiany w działaniu modelu w czasie.

Jednym z ważnych problemów w systemach produkcyjnych jest tzw. **drift danych**. Oznacza to sytuację, w której rozkład danych wejściowych zmienia się w czasie, co może prowadzić do pogorszenia jakości modelu.

Analiza ważności cech i innych wyjaśnień pozwala zauważyć, że model zaczyna opierać swoje decyzje na innych czynnikach niż wcześniej.

---

### Regulacje i etyka

Ostatnią ważną grupą interesariuszy są **instytucje regulacyjne oraz audytorzy**.

W wielu dziedzinach wykorzystanie algorytmów podlega regulacjom prawnym. Przykładem są systemy podejmujące decyzje o charakterze finansowym, medycznym lub administracyjnym.

W takich przypadkach wyjaśnialność modeli ma kilka kluczowych funkcji:

* umożliwia przeprowadzenie audytu systemu,
* zwiększa transparentność działania organizacji,
* zmniejsza ryzyko szkody dla użytkowników,
* pozwala wykryć potencjalną dyskryminację.

Z perspektywy etycznej wyjaśnienia są również ważne dlatego, że umożliwiają **rozliczalność systemów sztucznej inteligencji**. Jeśli decyzja modelu okaże się błędna lub szkodliwa, konieczne jest zrozumienie, dlaczego system ją podjął.

# 5. Przykłady zastosowań XAI

Wyjaśnialność modeli jest szczególnie ważna w obszarach, w których decyzje mają duże znaczenie społeczne lub ekonomiczne.

W medycynie systemy sztucznej inteligencji są wykorzystywane do analizy obrazów diagnostycznych, takich jak zdjęcia rentgenowskie, tomografia komputerowa czy obrazy z rezonansu magnetycznego. Model może wskazać obszary obrazu, które sugerują obecność choroby. Lekarz powinien jednak wiedzieć, czy model rzeczywiście analizuje istotne elementy obrazu, czy może reaguje na przypadkowe artefakty.

W bankowości modele predykcyjne służą do oceny ryzyka kredytowego. Jeśli system odrzuca wniosek kredytowy, klient powinien mieć możliwość uzyskania informacji, które czynniki miały największy wpływ na tę decyzję.

W przypadku autonomicznych pojazdów systemy sztucznej inteligencji podejmują decyzje dotyczące hamowania, skręcania czy przyspieszania. W sytuacji wypadku niezwykle ważne jest zrozumienie, dlaczego system podjął określoną decyzję.

Innym przykładem są systemy wspomagające decyzje w wymiarze sprawiedliwości. W niektórych krajach wykorzystuje się modele oceniające ryzyko ponownego popełnienia przestępstwa przez skazaną osobę. W takich przypadkach szczególnie ważne jest, aby decyzje podejmowane przez system były przejrzyste i sprawiedliwe.

# 6. Zaufanie do systemów AI – aspekt psychologiczny

Wyjaśnialność modeli ma również ważny wymiar psychologiczny. Ludzie mają naturalną potrzebę rozumienia decyzji podejmowanych przez systemy, z których korzystają. Nawet jeśli model jest bardzo dokładny, brak wyjaśnienia może prowadzić do braku zaufania.

Badania z zakresu psychologii poznawczej pokazują, że ludzie są bardziej skłonni zaufać systemowi, jeśli otrzymują zrozumiałe wyjaśnienie jego działania. Jednocześnie wyjaśnienia nie mogą być zbyt skomplikowane. Jeśli system generuje bardzo złożone raporty techniczne, użytkownicy mogą je ignorować lub uznać za niezrozumiałe.

Istnieje również zjawisko określane jako **automatyzacyjna stronniczość**. Polega ono na tym, że ludzie mają tendencję do bezkrytycznego akceptowania decyzji systemów komputerowych, szczególnie jeśli są one prezentowane jako bardzo dokładne. Wyjaśnienia mogą pomóc użytkownikom lepiej ocenić, czy decyzja modelu jest sensowna.

Kolejnym ważnym pytaniem jest kwestia odpowiedzialności. Jeśli decyzja została podjęta przez algorytm, kto ponosi odpowiedzialność za jej konsekwencje? Czy jest to programista, który stworzył model? Firma, która go wdrożyła? A może użytkownik, który zdecydował się z niego skorzystać? Wyjaśnialność modeli pomaga w analizie takich sytuacji, ponieważ umożliwia zrozumienie procesu decyzyjnego systemu.

---

# 7. Podsumowanie

Wyjaśnialna sztuczna inteligencja stanowi ważny obszar badań w nowoczesnym uczeniu maszynowym. Wraz z rosnącą złożonością modeli pojawia się potrzeba lepszego zrozumienia ich działania.

Modele uczenia maszynowego często działają jak czarne skrzynki – generują predykcje, ale nie zawsze potrafimy wyjaśnić ich logikę. XAI dostarcza narzędzi, które pozwalają analizować wpływ poszczególnych cech na decyzje modelu, wizualizować jego zachowanie oraz oceniać jakość wyjaśnień.

Wyjaśnialność jest ważna z wielu powodów: zwiększa bezpieczeństwo systemów, umożliwia spełnienie wymogów prawnych, pomaga wykrywać uprzedzenia w danych oraz wspiera proces debugowania modeli.

W kolejnych zajęciach poznamy konkretne metody wykorzystywane w XAI, takie jak analiza ważności cech, wykresy zależności częściowych, metoda LIME oraz metoda SHAP. Dzięki nim będziemy mogli zobaczyć, w jaki sposób można w praktyce analizować decyzje modeli uczenia maszynowego.
